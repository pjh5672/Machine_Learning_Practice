```python
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
```

### KMeans clustering

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
X = iris.data[:, :2]
y = iris.target
```

```python
X
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='gist_rainbow')
plt.xlabel('Sepal Length', fontsize=18)
plt.ylabel('Sepal Width', fontsize=18)
```

```python
km = KMeans(n_clusters=3, n_jobs=4, random_state=21)
km.fit(X)
```

```python
centers = km.cluster_centers_
new_labels = km.labels_
print(centers)
```

```python
#     
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(X[:,0], X[:,1], c=y, cmap='gist_rainbow', edgecolor='k', s=150)
axes[0].set_xlabel('Sepal length', fontsize=15)
axes[0].set_ylabel('Sepal width', fontsize=15)
axes[0].tick_params(direction='in', length=5, width=1, colors='k', labelsize=15)
axes[0].set_title('Actual', fontsize=18)

axes[1].scatter(X[:,0], X[:,1], c=new_labels, cmap='jet', edgecolor='k', s=150)
axes[1].set_xlabel('Sepal length', fontsize=15)
axes[1].set_ylabel('Sepal width', fontsize=15)
axes[1].tick_params(direction='in', length=5, width=1, colors='k', labelsize=15)
axes[1].set_title('Predicted', fontsize=18)
```

### 2  Kmeans clustering

```python
from sklearn.datasets import make_blobs

# 
X, y = make_blobs(n_samples=150, n_features=2, centers=3, cluster_std=0.5, shuffle=True, random_state=0)
```

```python
# 
plt.scatter(X[:,0], X[:,1], c='white', marker='o', edgecolors='black', s=50)
plt.show()
```

```python
km = KMeans(n_clusters=3, init='random', n_init=10, max_iter=300, tol=1e-04, random_state=0)
y_km = km.fit_predict(X)
```

```python
y_km
```

```python
#   
plt.scatter(X[y_km==0, 0], X[y_km==0, 1], s=50, c='lightgreen', marker='s', edgecolors='black', label='cluster 1')
plt.scatter(X[y_km==1, 0], X[y_km==1, 1], s=50, c='orange', marker='o', edgecolors='black', label='cluster 2')
plt.scatter(X[y_km==2, 0], X[y_km==2, 1], s=50, c='lightblue', marker='v', edgecolors='black', label='cluster 3')

#   
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s=250, c='red', marker='*', edgecolors='black', label='centroid')

plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

###  K  

```python
# Elbow method
distortions = []

for i in range(1, 11):
    km = KMeans(n_clusters=i, init='random', n_init=10, max_iter=300, tol=1e-04, random_state=0)
    km.fit(X)
    distortions.append(km.inertia_) 
    
# plt
plt.plot(range(1,11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
"""
"km.inertia_"      ,       K    K .
,  k=3       .
"""
```

###  Kmeans clustering   
-   , Kmeans clustering  

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score

documents = ["This little kitty came to play when I was eating at a restaurant.","hello kitty is my favorite character",
             "Merley has the best squooshy kitten belly.","Is Google translator so good?","google google"
             "google Translate app is incredible.","My dog s name is Kong","dog dog dog","cat cat"
             "If you open 100 tab in google you get a smiley face.","Kong is a very cute and lovely dog",
             "Best cat photo I've ever taken.","This is a cat house"
             "Climbing ninja cat kitty.","What's your dog's name?","Cat s paws look like jelly",
             "Impressed with google map feedback.","I want to join google","You have to wear a collar when you walk the dog",
             "Key promoter extension for google Chrome.","Google is the best company","Google researcher"]

vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(documents) # document   (,   )
```

```python
X # 55 19 
```

```python
true_k = 3
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X)
```

```python
order_centroids = model.cluster_centers_.argsort()[:, ::-1] # cluster_centers_     index  
terms = vectorizer.get_feature_names()
```
