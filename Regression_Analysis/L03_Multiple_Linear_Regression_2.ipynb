```python
import os
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
boston = pd.read_csv('./data/Boston_house.csv')
boston.head()
```

```python
x_data = boston[['CRIM', 'RM', 'LSTAT']]
target = boston[['Target']]
x_data.head()
```

```python
x_data1 = sm.add_constant(x_data, has_constant='add')
```

```python
multi_model = sm.OLS(target, x_data1)
fitted_multi_model = multi_model.fit()
```

```python
fitted_multi_model.summary()
```

**VIF를 통한 다중공선성 확인**

```python
"""
[ VIF의 의미 ] 
정해진 변수를 종속변수로 두고, 나머지 변수들이 종속변수에 기여하는 정도를 회귀분석했을때 변동성 비율(R-squared)를 1에서 뺀 역수를 
의미해서, 10 이상일 경우 해당변수가 다중공선성이 있다고 의미함
"""

from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['feature'] = x_data.columns
vif['VIF Factor'] = [variance_inflation_factor(x_data.values, i) for i in range(x_data.shape[1])]
vif
```

```python
# nox 변수 제거 후 (x_data3) VIF 확인
vif = pd.DataFrame()
x_data3 = x_data.drop('NOX', axis=1)
vif['feature'] = x_data3.columns
vif['VIF Factor'] = [variance_inflation_factor(x_data3.values, i) for i in range(x_data3.shape[1])]
vif
```

```python
# RM 변수 제거 후 (x_data4) VIF 확인
vif = pd.DataFrame()
x_data4 = x_data3.drop('RM', axis=1)
vif['feature'] = x_data4.columns
vif['VIF Factor'] = [variance_inflation_factor(x_data4.values, i) for i in range(x_data4.shape[1])]
vif
```

**nox 변수 제거한 데이터(x_data3) 상수항 추가 후 회귀모델 적합**

```python
x_data3_ = sm.add_constant(x_data3, has_constant='add')
# 회귀모델 적합
multi_model3 = sm.OLS(target, x_data3_)
fitted_multi_model3 = multi_model3.fit()
fitted_multi_model3.summary()
```

**nox, rm 변수 제거한 데이터(x_data4) 상수항 추가 후 회귀모델 적합**

```python
x_data4_ = sm.add_constant(x_data4, has_constant='add')
# 회귀모델 적합
multi_model4 = sm.OLS(target, x_data4_)
fitted_multi_model4 = multi_model4.fit()
fitted_multi_model4.summary()
```

```python
"""
[ 회귀모델 결과 비교 ] 
- NOX 제거한 모델 : 종속변수(Y)에 대한 변동성이 66%(R-squared)로서 모든 변수를 사용해 적합시킨 모델과 차이가 없음. 
즉, NOX의 영향이 미미하다
- NOX, RM 제거한 모델 : 종속변수(Y)에 대한 변동성이 57%(R-squared)로서 66%에서 상당히 감소하였음. 
즉, RM은 유의미한 데이터로써 변수를 제거하지 않고 사용하는게 효과적임.
"""
```

**학습/검증데이터 분할**

```python
from sklearn.model_selection import train_test_split
X = sm.add_constant(x_data, has_constant='add')
Y = target
train_x, test_x, train_y, test_y = train_test_split(X, Y, train_size=0.7, test_size=0.3, random_state=1)
print(train_x.shape, test_x.shape, train_y.shape, test_y.shape)
```

```python
# 회귀모델 적합
fit_model = sm.OLS(train_y, train_x)
fitted_model = fit_model.fit()
fitted_model.summary()
```

```python
# 검증데이터에 대한 예측값과 true 값 비교
plt.plot(np.array(fitted_model.predict(test_x)), label='pred')
plt.plot(np.array(test_y), label='true')
plt.legend()
plt.show()
```

**NOX변수만 제거한 데이터에 적합시켜 예측한 모델**

```python
X = sm.add_constant(x_data3, has_constant='add')
Y = target
train_x, nox_test_x, train_y, nox_test_y = train_test_split(X, Y, train_size=0.7, test_size=0.3, random_state=1)
print(train_x.shape, nox_test_x.shape, train_y.shape, nox_test_y.shape)

# 회귀모델 적합
nox_model = sm.OLS(train_y, train_x)
nox_fitted_model = nox_model.fit()
```

**NOX, RM 변수 제거한 데이터에 적합시켜 예측한 모델**

```python
X = sm.add_constant(x_data4, has_constant='add')
Y = target
train_x, noxrm_test_x, train_y, noxrm_test_y = train_test_split(X, Y, train_size=0.7, test_size=0.3, random_state=1)
print(train_x.shape, noxrm_test_x.shape, train_y.shape, noxrm_test_y.shape)

# 회귀모델 적합
noxrm_fit_model = sm.OLS(train_y, train_x)
noxrm_fitted_model = noxrm_fit_model.fit()
```

```python
# 검증데이터에 대한 예측값과 true 값 비교
plt.plot(np.array(nox_fitted_model.predict(nox_test_x)), label='nox')
plt.plot(np.array(noxrm_fitted_model.predict(noxrm_test_x)), label='nox_rm')
plt.plot(np.array(test_y), label='true')
plt.legend()
plt.show()
# NOX만 제거한 모델이 NOX와 RM을 제거한 적합모델보다 더 잔차정확도가 높음
```

**MSE를 통한 검증데이터 성능비교**

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_true=test_y['Target'], y_pred=fitted_model.predict(test_x))
```

```python
mean_squared_error(y_true=test_y['Target'], y_pred=nox_fitted_model.predict(nox_test_x))
```

```python
mean_squared_error(y_true=test_y['Target'], y_pred=noxrm_fitted_model.predict(noxrm_test_x))
```
